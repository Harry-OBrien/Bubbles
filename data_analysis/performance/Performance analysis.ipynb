```python
import matplotlib.pyplot as plt
import pandas
```

```python
dataframes = []
for i in range(4, 14, 2):
    file_name = "./framerate_at_res_{}.csv".format(i)
    print(file_name)
    df = pandas.read_csv(file_name)
    dataframes.append(df)
```

```python
dataframes[0].columns
```

```python
fig, ax = plt.subplots()  # Create a figure and an axes.
for i, df in enumerate(dataframes):
    ax.plot(df["fps"], label="resolution {}".format(i*2 + 4))
ax.set_title("A plot to show how a particle system performs at different resolution values.")  # Add a title to the axes.

ax.legend()  # Add a legend.
```

```python
fig, ax1 = plt.subplots()  # Create a figure and an axes.
df = pandas.read_csv("framerate_with_increasing_particles .csv")
color='tab:blue'
ax1.plot(df["fps"], label="fps")
ax1.set_xlabel('frame')
ax1.set_ylabel('fps', color=color)
ax1.set_title("A plot to show the fps as particle count increases")  # Add a title to the axes.
ax1.tick_params(axis='y', labelcolor=color)

color = 'tab:red'
ax2 = ax1.twinx()
ax2.plot(df["particle_count"], label="particle_count", color=color)
ax2.set_ylabel("particles", color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax.legend()  # Add a legend.
```
